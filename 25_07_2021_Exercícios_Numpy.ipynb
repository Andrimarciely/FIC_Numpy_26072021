```python
import numpy as np

def print_matrix(matrix):
  print('\n'.join([''.join(['{:4}'.format(item) for item in row]) for row in matrix]))


def average_and_standard_deviation(matrix):
  average_row = []
  average_row = [np.median(row) for row in matrix]

  standard_deviation_row = []
  standard_deviation_row = [np.std(row) for row in matrix]
  
  average_col = [] 
  average_col = [np.median(col) for col in matrix]
  
  standard_deviation_col = []
  standard_deviation_col = [np.std(col) for col in matrix]
 
  print("Média linhas matriz → ", average_row)
  print("Desvio padrão linhas matriz → ", standard_deviation_row)

  print("Média linhas matriz → ", average_col)
  print("Desvio padrão colunas matriz → ", standard_deviation_col)




if __name__ == "__main__":
  ExA = [[12,9,4,1],[11,5,8,1],[1,2,3,1]]
  print('ExA')
  print_matrix(ExA)
  print('\n')

  ExB = [[1,5],[1,7],[1,9],[1,1]]
  print('ExB')
  print_matrix(ExB)
  print('\n')

  ExC = np.random.randint(0,100,(3,3))
  print('ExB')
  print_matrix(ExC)
  print('\n')
  print('\nDados matriz ExA')
  average_and_standard_deviation(ExA)
  print('\nDados matriz ExC')
  average_and_standard_deviation(ExC)
  print('\n')


  ExA = np.matrix(ExA)
  ExD = []
  ExD = ExA[:,-2:]
  print('\nMatriz ExD')
  print(ExD)
  print('\n')
  print('\nMatriz ExE')
  p1 = ExA[0:2,]
  print('P1\n',p1)
  p2 = np.transpose(ExA[:,0:2])
  print('P2\n',p2)
  Exe = np.concatenate((p1,p2),1)
  print('\nMatriz ExE')
  print(ExE)
```

```python
'''Função da validação de entradas'''

from typing import Any, Callable

def receive_value(type_value: Callable, message: str) -> Any:

    while True:
        value = input(f"{message} → ")

        try:
            return type_value(value)

        except ValueError:
            print(f"'{value}' não é válido, tente novamente.")
```
